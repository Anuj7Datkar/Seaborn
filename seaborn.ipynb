```python
# the ways of styling themes are as follows:
# white
# dark
# whitegrid
# darkgrid
# ticks

# set_context() allows us to override default parameters.  this affects things like the size of the labels, lines, and other elements of the plot,  but not the overall style -> set_setyle(). 
# The context are:
# poster - suitable for large displays or posters
# paper - ideal for printing in academic papers
# notebook - good for working inside Jupyter notebooks (default setting)
# talk - good for presentation slides where readability matters

# palette -  colors styling :
# Set1 - bright and high-contrast colors
# Set2 - softer,more pastel-like colors
# Set3 - a mix of soft and bold, varied hues

# palette parameters:
# deep - default pallete,rich and balanced
# muted - softer colors,good for many cateogories
# bright - high-contrast, bold colors
# pastel - soft,pale tones
# dark - darker tones, easy to distinguis
# colorblind - optimized for colorblind accessibility

# palette type:
# qualitative - categorical in nature, the color assigned to each group need to be distinct
# sequential - data with ordered progression(low -> high)
# diverging - data with meaningful midpoint or mixed value like +ve and -ve
```

```python
sns.set_style("ticks")
sns.set_context("notebook")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue ="island", palette = "Dark2")
sns.despine(left = True)
```

```python
sns.scatterplot(data = penguins, x ="species", y = "body_mass_g", hue = "island", style = "sex", alpha = 0.5)
```

```python
sns.stripplot(data = penguins, x ="species", y = "body_mass_g", hue = "island",dodge = True, jitter =True)
```

```python
# swarm plot
sns.set_context("talk")
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g", hue = "island")
sns.despine()
```

```python
# histogram
sns.set_context("notebook")
sns.histplot(data = penguins, x = "body_mass_g", hue = "sex",multiple = "stack" )
```

```python
# regression plot:
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", color ="green")
```

```python
# line plot:
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue = "island", style = "sex")
```

```python
#joint plot
sns.jointplot(data = penguins, x = "body_mass_g",y = "flipper_length_mm")
```

```python
sns.jointplot(data = penguins, x = "body_mass_g",y = "flipper_length_mm", hue = "sex")
```

```python
# bar plot:
sns.barplot(data = penguins, x = "species",y = "body_mass_g", hue = "sex", palette = ["pink","skyblue"], estimator = np.sum)
```

```python
# count plot:
sns.countplot(data = penguins, x = "species", hue = "island", palette = "Set2")
```

```python
#box plot:
sns.boxplot(data = penguins, x = "species",y = "body_mass_g", hue = "sex")
```

```python
# violin plot:
sns.set_style("whitegrid")
sns.violinplot(data = penguins, x = "species",y = "body_mass_g", hue = "sex", split = True, inner = "box" )
```

```python
# width of violin = density of the data at different values
# middle line = median
# thick bar = interquartile range (if inner="box")
```

```python
sns.violinplot(data = penguins,x = "species",y = "body_mass_g")
sns.swarmplot(data = penguins,x = "species",y = "body_mass_g",hue = "island", size = 3,palette = "dark")
```

```python
#KDE plot(kernel density estimation): for estimation the probability function of continuous or non-parametric data
# OR in simple terms: a smooth curve that shows the distribution of data - like a smoothed-out version of histogram
# It helpsyou see where the data is concentrated (peaks) and how it spreads


sns.kdeplot(data = penguins,x = "body_mass_g",hue = "species",palette = "pastel", fill = "True"                                     )
```

```python
# heatmap: a 2D color-coded matrix that shows relationships or patterns between two variables.penguins
# It's commonly used to :
#  - show correlation between features
#  - visualize confusion matrices -> will see in the later stages
#  - Highlight missing data ->sns.heatmap(df.isnull())

penguins[columns].corr()
```

```python
columns = ["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]

sns.heatmap(data = penguins[columns].corr(),annot = True, cmap = "Reds",vmin = -0.2,linewidth = "2",linecolor = "black")
```

```python
# Rug plot:

sns.rugplot(data = penguins,y ="body_mass_g",hue = "species",palette ="Set2", height = 0.2)
```

```python
#Pair plot:

sns.pairplot(data = penguins, hue = "sex",palette = "Set2",diag_kind = "hist")
```

```python
# pair grid: subplot grid for ploting pairwise relationships in a dataset

graph = sns.PairGrid(data = penguins,hue ="sex",palette ="Set2")
graph.map_upper(sns.scatterplot) # upper triangle
graph.map_lower(sns.kdeplot) # lower triangle
graph.map_diag(sns.histplot) # diagonal
graph.add_legend()
```
